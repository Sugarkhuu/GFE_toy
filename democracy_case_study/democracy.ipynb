```python
## Case study
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# sns.set_theme()


df = pd.read_stata(r'5yearpanel.dta')
df_gfe = pd.read_stata(r'5yearpanel_GFE.dta')
df_gfe = df_gfe[['fhpolrigaug','lrgdpch','year','assignment','code']]
df_gfe['group'] = df_gfe['assignment'].map({1:'early transition',2:'low democracy',
                        3:'late transition',4:'high democracy'})

palette = {"high democracy":"tab:blue",
           "early transition":"black", 
           "late transition":"yellowgreen",
           "low democracy":"tab:brown",
           "e":"tab:orange", 
           "f":"tab:purple"}
```

```python
df_gfe
```

```python
plt.figure()
sns.lineplot(data=df_gfe,x='year',y='fhpolrigaug', hue='group',
        hue_order = ['high democracy', 'early transition', 'late transition','low democracy'],palette=palette)
plt.legend(loc='lower center', ncol=4)
plt.ylabel('Democracy')
plt.xlabel(None)
plt.savefig('democracy.png')
```

```python
plt.figure()
sns.lineplot(data=df_gfe,x='year',y='lrgdpch', hue='group',
        hue_order = ['high democracy', 'early transition', 'late transition','low democracy'],palette=palette)
plt.legend(loc='lower center', ncol=4)
plt.ylabel('log GDP per capita')
plt.xlabel(None)
plt.savefig('gdp.png')
# plt.show()
```

```python
x = np.linspace(0,10,21)
g1 = np.sin(x/4)
g2 = x**2/100
plt.plot(g1)
plt.plot(g2)
```

```python
np.random.seed(seed=5)
scale = 0.1
g11 = g1 + np.random.normal(scale=scale,size=len(g1))
g12 = g1 + np.random.normal(scale=scale,size=len(g1))
g13 = g1 + np.random.normal(scale=scale,size=len(g1))
g21 = g2 + np.random.normal(scale=scale,size=len(g1))
g22 = g2 + np.random.normal(scale=scale,size=len(g1))

plt.scatter(x,g11,c='b')
plt.scatter(x,g12,c='k')
plt.scatter(x,g13,c='y')
plt.scatter(x,g21,marker='*')
plt.scatter(x,g22,marker='*')
plt.xlabel('Time')
plt.ylabel('Y - variable of interest')
```
